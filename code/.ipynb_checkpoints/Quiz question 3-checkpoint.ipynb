```python
from gensim.models import Word2Vec
import PyPDF2

head=['Education','Skills','Research', 'Professional']
keep=[]

os.chdir('code') #if you are in Quiz2 folder

#if you are in code folder then please comment the line 7.

reader = PyPDF2.PdfFileReader("Sourin_Dey_Resume_.pdf")
x=[]
for page in reader.pages:
    text = page.extractText()
    x.append(text)
# x=x[0]+x[1]
for it in range(len(x)):
    for x11,x12 in enumerate(x[it]):
        if x12.isspace():
            keep.append(x11)
        

c1=0
a=0

c2=keep[a]
vocab=[]
word=x[0][c1:c2]
# print(word)
counter=0
for e in range(len(keep)):
    try:
        c1=keep[a]
        c2=keep[a+1]
        word=x[0][c1:c2]
    #     if word in head[0]:
    #         break
    #     print(word)
        a1=word.strip('\n')
        vocab.append(a1)
    #     if a1==head[1]:
    #         break
        a+=1
    except:
        print('second page came up')
    
head=['Education','Skills','Research', 'Professional','Projects','Publications']
store=[]
dup=[]
for x2,x1 in enumerate(vocab):
    x1=x1.strip(' ')
#     print(x)
    if x1 in head:
        if x1 not in dup:
            dup.append(x1)
            store.append(x2)
        
for x4 in store:
    print(vocab[x4])
print('\n'*5)    
    
resume=[]
# resume_dictionary={'Education':[], 'Skills':[], 'Research':[], 'Professional':[]}
for i4 in range(len(store)):
    print(dup[i4])
    try:
        resume.append(vocab[store[i4]:store[i4+1]])
        print(vocab[store[i4]:store[i4+1]])
        print('\n'*4)
    except:
        resume.append(vocab[store[i4]:])
        print(vocab[store[i4]:])
        print("last ")
```

```python
corpus=resume
# train model
model = Word2Vec(corpus, min_count=1)
# summarize the loaded model
print("INFO: Model - \n" + str(model))

words = list(model.wv.index_to_key)

# remove words and symbols like :the, to, -,: and so on
vocab2=[i for i in words if len(i)>3]
vocab2

import pandas as pd
df=pd.DataFrame(columns=['col_'+str(i1) for i1 in range(101)])

for x5,x6 in enumerate(vocab2):
    vals=model.wv[x6]
    df.loc[x5,'col_0']=x6
    df.loc[x5,df.columns[1:]]=[i7 for i7 in vals]
```

```python
df2=df.transpose()
df2.columns=df2.iloc[0,:]
df2=df2.iloc[1:,:]
df2=df2.reset_index(drop=True)
# df2=df2.drop('col_0',axis=1)
# print(df2.head(2))
print('\n'*5)
print('Embedding Statistics')
print('---'*25)
print("Name....Mean....Max....Min")
print('---'*25)
for x12,x13 in enumerate(df2.columns):
    print(x13,round(df2[x13].mean(),6),round(df2[x13].max(),6),round(df2[x13].min(),6))
```

```python
df2=pd.read_fwf('projs.txt',header=None)
corpus2=[[i8] for i8 in df2[df2.columns[0]]]
corpus_final=[]
for i9 in corpus2:
    corpus_final.append(i9[0].split())
```

```python
model = Word2Vec(corpus_final, min_count=1)
words = list(model.wv.index_to_key)
vocab2=[i for i in words if len(i)>3]

vocab2
```

```python
import pandas as pd
df=pd.DataFrame(columns=['col_'+str(i1) for i1 in range(101)])

for x5,x6 in enumerate(vocab2):
    vals=model.wv[x6]
    df.loc[x5,'col_0']=x6
    df.loc[x5,df.columns[1:]]=[i7 for i7 in vals]
```

```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib import pyplot
plt.figure(figsize=(18,14))

X=[model.wv[x6] for x6 in vocab2]
pca = PCA(n_components=2)
result = pca.fit_transform(X)
# create a scatter plot of the projection
pyplot.scatter(result[:, 0], result[:, 1])
words = list(model.wv.index_to_key)
for i, word in enumerate(words):
     if i==96:break
     pyplot.annotate(word, xy=(result[i, 0], result[i, 1]))
    
os.chdir('..')
os.chdir('data')
plt.savefig('question_3c.png',dpi=300)
os.chdir('..')
os.chdir('code')
pyplot.show()
```
